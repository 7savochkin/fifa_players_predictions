```python
# Read dataframe
import pandas as pd
import numpy as np

df = pd.read_csv('../files/fifa_players.csv', low_memory=False)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
```

```python
# First of all, lets do EDA. 
# Exploratory Data Analysis is a crucial step that involves analyzing datasets to summarize their main characteristics,
def EDA(df):
    total_na = df.isna().sum().sum()
    print(f"Dimensions {df.shape[0]} rows, {df.shape[1]} cols")
    print(f"Total NA's: {total_na}")
    print("Column Type Unique NA")
    col_name = df.columns
    dtypes = df.dtypes
    uniq = df.nunique()
    na_val = df.isna().sum()
    for i in range(len(df.columns)):
        print(f"{col_name[i]} {dtypes[i]} {uniq[i]} {na_val[i]}")

# Now, we can see all columns and their characteristics (Type of columns, Count of unique values and count of NA values)
EDA(df)
```

```python
# Lets see also describing statistic of out dataframe
df.describe()
```

```python
# As we know, our task is classifing the player's position, 
# and we may omit some columns, which is not useful.
# First of all, we drop fifa id, short and long names, because name doesn't involve on position.
# We drop info about club and league, because league is not useful for us and for club, we have club_team_id.
# Finally, all urls and others uninformative columns were dropped.
dropped_columns = ["sofifa_id", "player_url", "short_name", "long_name", "club_name", "league_name", "league_level",
                   "club_loaned_from", "club_joined", "nation_jersey_number", "nation_team_id", "release_clause_eur", "player_tags", "player_traits",
                   "player_face_url", "nation_logo_url", "nation_flag_url", "club_logo_url", "dob", "club_jersey_number", "value_eur", 
                   "wage_eur", "club_contract_valid_until", "club_team_id"]
# To remove other uninformative columns, we need to see the correlation

clear_df = df.drop(columns=dropped_columns)
```

```python
# Also we drop NA in club_position
# And set Without to nation_position
clear_df = clear_df.dropna(subset=['club_position'])
clear_df['nation_position'] = clear_df['nation_position'].fillna('Without')
```

```python
# !!!ATENTION!!!
# U should uncomment this code if u wanna drop club_position and nation_position
clear_df.drop(columns=['club_position', 'nation_position'], axis=1, inplace=True)
```

```python
# After this we need to find all categorical and numerical features, so filter them by dtype
categorical_features = [col for col in clear_df.columns if clear_df[col].dtype == "O"]
numerical_features = [col for col in clear_df.columns if clear_df[col].dtype in ["int64", "float64"]]
```

```python
# I set mean value to float columns where is NA
columns_float_to_fill_na = ["pace", "shooting", "passing", "dribbling", "defending", "physic", "goalkeeping_speed"]
for column in columns_float_to_fill_na:
    clear_df[column] = clear_df[column].fillna(clear_df[column].mean())
```

```python
# After this we need to see updated EDA
EDA(clear_df)
```

```python
# Clear outliers
# IQR
lower_boundaries = []
upper_boundaries = []

def replace_outliers_for_columns(df: pd.DataFrame, columns: list) -> None:
    for column in numerical_features:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
    
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
    
        lower_boundaries.append(lower_bound)
        upper_boundaries.append(upper_bound)
    
        df[column] = np.where(df[column] > upper_bound, 
                                      upper_bound, df[column])
        df[column] = np.where(df[column] < lower_bound,
                                      lower_bound, df[column])
    
        print(column, "--", lower_bound, "--", upper_bound)

replace_outliers_for_columns(clear_df, numerical_features)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer, LabelEncoder, OneHotEncoder

# So we have 3 types of columns which will be encoded differently.
# 1. Player_positions can be array with multiple positions 
# so we split to array and use MultiLabelBinarizer for multiclassess. And encoding with OneHotEncoder.
# 2. Club_position,  will be encoded with OneHotEncoding
# 3. All other will be encoded with LabelEncoder, because we have many nationalities.

def encode_multiclassess(df: pd.DataFrame, column: str) -> pd.DataFrame:
    df[column] = df[column].apply(lambda x: x.split(', '))

    mlb = MultiLabelBinarizer()
    mlb_encoded = mlb.fit_transform(df[column])
    return pd.DataFrame(mlb_encoded, columns=mlb.classes_)


def encode_label_encoder(df: pd.DataFrame, column: str) -> pd.DataFrame:
    le = LabelEncoder()
    le_encoded = le.fit_transform(df[column])
    return le_encoded

def encode_one_hot_encoder(df: pd.DataFrame, columns: list) -> pd.DataFrame:
    # ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')
    ohe = OneHotEncoder(sparse_output=False)
    ohe_encoded = ohe.fit_transform(df[columns])
    return pd.DataFrame(ohe_encoded, columns=ohe.get_feature_names_out(columns))

categorical_special_cols = ("player_positions", "nationality_name")
categorical_onehotencoding_cols = list(filter(lambda x: x not in categorical_special_cols, categorical_features))
player_position_col, nationality_name_col = categorical_special_cols

player_positions_encoded_df = encode_multiclassess(clear_df, player_position_col)
nationality_name_encoded_df = encode_label_encoder(clear_df, nationality_name_col)
categorical_onehotencoded_df = encode_one_hot_encoder(clear_df, categorical_onehotencoding_cols)
```

```python
# for label encoder we only change column values
clear_df[nationality_name_col] = nationality_name_encoded_df

print(clear_df.shape)

# for OneHotEncoder we merge our main df and encoded columns df
clear_df = clear_df.reset_index(drop=True)
df_encoded = pd.concat([clear_df, categorical_onehotencoded_df], axis=1)
print(df_encoded.shape)
```

```python
# After encoding we can get X and y dataframes
X = df_encoded.drop(columns=[player_position_col, *categorical_onehotencoding_cols], axis=1)
y = player_positions_encoded_df
```

```python
# Build train, validation and test dfs
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.1, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import PolynomialFeatures

# find all interactions and double them
poly = PolynomialFeatures(degree=2, interaction_only=True)
X_train_poly = poly.fit_transform(X_train)
X_valid_poly = poly.fit_transform(X_valid)
X_test_poly = poly.fit_transform(X_test)
```

```python
"""
I use Mutual Information Classifier because:
- It captures non-linear relationships between the features and the labels, 
  which many other methods (like correlation or linear models) might miss. 
- Handles Multi-Label Structure
Mutual Information (MI) tells us how much knowing one variable reduces uncertainty about another variable.
It is particularly useful for detecting both linear and non-linear dependencies between variables.
"""

from sklearn.feature_selection import RFE, mutual_info_classif
from sklearn.ensemble import RandomForestClassifier

y_train_categorical = np.argmax(y_train, axis=1)

mi = mutual_info_classif(X_train, y_train_categorical)
mi_df = pd.DataFrame({'Feature': X_train.columns, 
                      'Mutual Information': mi}).sort_values(by='Mutual Information', ascending=False)

# Create a list of features with mutual information >= 0.1
selected_features = mi_df[mi_df['Mutual Information'] >= 0.1]['Feature'].tolist()

# Filter X_train, X_test, and X_valid
X_train_mi = X_train[selected_features]
X_valid_mi = X_valid[selected_features]
X_test_mi = X_test[selected_features]

# Optionally, display the shapes of the filtered datasets
print("Filtered X_train shape:", X_train_mi.shape)
print("Filtered X_valid shape:", X_valid_mi.shape)
print("Filtered X_test shape:", X_test_mi.shape)
```

```python
# Build pipelines for different models for finding best model


from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from imblearn.ensemble import EasyEnsembleClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.multiclass import OneVsRestClassifier
from sklearn.multioutput import MultiOutputClassifier
from skmultilearn.adapt import MLkNN
from sklearn.pipeline import make_pipeline


def logreg_pipeline(X_train, y_train):
    'Logistic Regression'
    logreg_pipeline = make_pipeline(StandardScaler(), MultiOutputClassifier(LogisticRegression()))
    logreg_pipeline.fit(X_train, y_train)
    return logreg_pipeline


def rf_pipeline(X_train, y_train):
    'Random Forest'
    rf_pipeline = make_pipeline(StandardScaler(), MultiOutputClassifier(RandomForestClassifier()))
    rf_pipeline.fit(X_train, y_train)
    return rf_pipeline


def xgb_pipeline(X_train, y_train):
    'XGBoost'
    xgb_pipeline = make_pipeline(StandardScaler(), MultiOutputClassifier(XGBClassifier()))
    xgb_pipeline.fit(X_train, y_train)
    return xgb_pipeline


def easy_ensemble_pipeline(X_train, y_train):
    'Easy Ensemble Classifier'
    easy_ensemble = MultiOutputClassifier(EasyEnsembleClassifier())
    easy_ensemble.fit(X_train, y_train)
    return easy_ensemble
```

```python
pipelines = {
    'Logistic Regression': logreg_pipeline,
    'Random Forest': rf_pipeline,
    'XGBoost': xgb_pipeline,
    'Easy Ensemble Classifier': easy_ensemble_pipeline,
}
```

```python
models = {}
for model_name, model_pipeline in pipelines.items():
    print(f"Training {model_name}...")
    models[model_name] = model_pipeline(X_train_mi, y_train)
```

```python
from sklearn.metrics import accuracy_score, hamming_loss, classification_report, jaccard_score, f1_score

def evaluate_models(model, X_test, y_test):
    y_pred = model.predict(X_test)
    
    subset_acc = accuracy_score(y_test, y_pred)
    print(f'Subset Accuracy (Exact Match Ratio): {subset_acc}')
    
    # Hamming Loss
    hamming = hamming_loss(y_test, y_pred)
    print(f'Hamming Loss: {hamming}')
    
    # Classification Report (Precision, Recall, F1 Score)
    print('Classification Report:')
    print(classification_report(y_test, y_pred))
    
    # Jaccard Similarity Score
    jaccard = jaccard_score(y_test, y_pred, average='samples')
    print(f'Jaccard Similarity Score: {jaccard}')
    
    # F1 Score (Micro and Macro)
    f1_micro = f1_score(y_test, y_pred, average='micro')
    f1_macro = f1_score(y_test, y_pred, average='macro')
    print(f'F1 Score (Micro): {f1_micro}')
    print(f'F1 Score (Macro): {f1_macro}')
```

```python
for model_name, model in models.items():
    print(f'Evaluation {model_name}')
    evaluate_models(model, X_valid_mi.values, y_valid.values)
```
