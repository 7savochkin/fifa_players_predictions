```python
# Read dataframe
import pandas as pd
import numpy as np

df = pd.read_csv('../files/fifa_players.csv', low_memory=False)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
```

```python
# 11. –ü—Ä–∏–¥—É–º–∞—Ç–∏ –º–µ—Ç—Ä–∏–∫—É, —è–∫–∞ –æ—Ü—ñ–Ω—é–≤–∞—Ç–∏–º–µ —è–∫—ñ—Å—Ç—å –ø–µ–Ω–∞–ª—å—Ç–∏—Å—Ç–∞.

# attacking_penalty_kick

# I decide to use mentality_penalties, attacking_finishing, mentality_composure, power_shot_power, movement_reactions
# because they better describe the characteristics of a pentalist. 
# I add them up and find the average.
# Also I think that attacking_finishing and mentality_penalties are more important and i add weighting factor. I multiple important columns on 0.4.
attacking_penalty_kick = [
    'mentality_penalties', 'attacking_finishing', 'mentality_composure', 'power_shot_power', 'movement_reactions'
]

df['attacking_penalty_kick'] = (0.4 * df['attacking_finishing'] + 0.4 * df['mentality_penalties'] + 
                      0.2 * (df['mentality_composure'] + df['power_shot_power'] + df['movement_reactions'])
                      ) / 1.4 # Divide on sum of coefficient.
```

```python
# 12. –ü—Ä–∏–¥—É–º–∞—Ç–∏ –º–µ—Ç—Ä–∏–∫—É, —è–∫–∞ –æ—Ü—ñ–Ω—é–≤–∞—Ç–∏–º–µ —è–∫—ñ—Å—Ç—å –≤–æ—Ä–æ—Ç–∞—Ä—è, —è–∫–∏–π –≤—ñ–¥–±–∏–≤–∞—î –ø–µ–Ω–∞–ª—å—Ç—ñ.

# goalkeeper_penalty_save

# I decide to use goalkeeping_reflexes, goalkeeping_diving, goalkeeping_positioning, goalkeeping_handling,
# goalkeeping_speed, mentality_composure, movement_reactions
# as they better describe the goalkeeper's characteristics.
# I add them up and find the average.
# Also I think goalkeeping_reflexes and goalkeeping_diving are more important and I add a weight factor. I multiply the sum of the important columns by 0.4
goalkeeper_penalty_save = [
     'goalkeeping_reflexes', 'goalkeeping_diving', 'goalkeeping_positioning',
     'goalkeeping_handling', 'mentality_composure'
]

df['goalkeeper_penalty_save'] = (0.4 * df['goalkeeping_reflexes'] + 0.4 * df['goalkeeping_diving']
                                    + 0.2 * (df['goalkeeping_positioning'] + df['goalkeeping_handling'] 
                                             + df['mentality_composure'])
                                   ) / 1.4
```

```python
# 13. –ß–∏ —î –∫–ª—É–±–∏, –≤ —è–∫–∏—Ö –ø–µ–Ω–∞–ª—å—Ç–∏—Å—Ç–∏ —Å—É—Ç—Ç—î–≤–æ (–ø—Ä–∏–¥—É–º–∞—Ç–∏, —â–æ –∑–Ω–∞—á–∏—Ç—å ‚Äú—Å—É—Ç—Ç—î–≤–æ‚Äù) –∫—Ä–∞—â—ñ –∑–∞ –≤–æ—Ä–æ—Ç–∞—Ä—ñ–≤, —è–∫—ñ –≤—ñ–¥–±–∏–≤–∞—é—Ç—å? –ê –Ω–∞–≤–ø–∞–∫–∏?

def get_best_goalkeeper_of_clubs(df: pd.DataFrame):
    return df[df["club_position"] == "GK"].groupby(
        by="club_name", as_index=False).agg({"goalkeeper_penalty_save": "max"})


def get_best_penalty_kickers_of_clubs(df: pd.DataFrame):
    return df[df["club_position"] != "GK"].groupby(
        by="club_name", as_index=False).agg({"attacking_penalty_kick": "max"})


def check_clubs_penalty_statistic(df: pd.DataFrame):

    clear_df = df[["club_name", "club_position", "attacking_penalty_kick", "goalkeeper_penalty_save"]].dropna()

    # get main goalkeepers of clubs
    goalkeeper_club_df = get_best_goalkeeper_of_clubs(clear_df)

    # get main penalty kickers of clubs
    kicker_club_df = get_best_penalty_kickers_of_clubs(clear_df)

    filtered_df = pd.merge(kicker_club_df, goalkeeper_club_df)

    result = pd.DataFrame()

    # If rating of features are greater than another feature on 3 points it's significantly!.
    # First of all get clubs with better kickers
    result["clubs_with_better_kicker"] = filtered_df[filtered_df["attacking_penalty_kick"] - filtered_df["goalkeeper_penalty_save"] >= 3].count()
    # Get clubs with better savers
    result["clubs_with_better_saver"] = filtered_df[filtered_df["goalkeeper_penalty_save"] - filtered_df["attacking_penalty_kick"] >= 3].count()
    
    return result.reset_index(drop=True)
```

```python
clubs_penalty_statistic = check_clubs_penalty_statistic(df)

# Statistic is showed that ussualy Penalty kickers are better than goalkeepers. 474 kickers > 14 goalkeeper savers
clubs_penalty_statistic.head(1)
```

```python
# 14. –¢–æ–ø 10 –∫–ª—É–±—ñ–≤ –≤ —è–∫–∏—Ö –Ω–∞–π–∫—Ä–∞—â—ñ –ø–µ–Ω–∞–ª—å—Ç–∏—Å—Ç–∏. (1 –±–∞–ª)

def top_clubs_with_best_penalty_kickers(df: pd.DataFrame, n=30):
    clear_df = df[["club_name", "club_position", "attacking_penalty_kick"]].dropna()

    result = get_best_penalty_kickers_of_clubs(clear_df)

    return result.sort_values(by="attacking_penalty_kick", ascending=False).head(n).reset_index(drop=True)

clubs_with_best_penalty_kickers = top_clubs_with_best_penalty_kickers(df, n=10)
clubs_with_best_penalty_kickers
```

```python
# 15. –¢–æ–ø 10 –∫–ª—É–±—ñ–≤ –≤ —è–∫–∏—Ö –Ω–∞–π–∫—Ä–∞—â—ñ –≤–æ—Ä–æ—Ç–∞—Ä—ñ

def top_clubs_with_best_goalkeepers(df: pd.DataFrame, n=30):
    clear_df = df[["club_name", "club_position", "goalkeeper_penalty_save"]].dropna()

    result = get_best_goalkeeper_of_clubs(clear_df)

    return result.sort_values(by="goalkeeper_penalty_save", ascending=False).head(n).reset_index(drop=True)

clubs_with_best_goalkeepers = top_clubs_with_best_goalkeepers(df, n=10)
clubs_with_best_goalkeepers
```
